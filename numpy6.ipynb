```python
#rounding function
x = np.array([1,2,3.7,4.1,5.3,5.6])

#round
print(np.round(x))
#floor
print(np.floor(x))
#ceil
print(np.ceil(x))
#trunc
print(np.trunc(x))
```

```python
#statical_function

arr = np.array([2,3,4,5,6,7])
#mean
print(np.mean(arr))

# median
print(np.median(arr))

#std()
print(np.std(arr))

#  var()
print(np.var(arr))

#  min()
print(np.min(arr))

# max()
print(np.max(arr))

#  argmin()
print(np.argmin(arr))

#  argmax()
print(np.argmax(arr))


# sum()
print(np.sum(arr))

# percentile()
sal = np.array([2008,4500,6600,7000])
print(np.percentile(sal,90))
```

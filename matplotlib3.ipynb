```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
enrol = [1,2,3,4,5,6]
marks = [10,20,30,40,50,60]
df = pd.DataFrame([])
```

```python
vegetables = {
    "carrot" :40,
    "potato" :25,
    "tomato" :30,
    "spinach" :30
}
names = list(vegetables.keys())
prices = list(vegetables.values())
```

```python
plt.figure(figsize = (5,5))
plt.plot(names,prices)
plt.xlabel("names")
plt.ylabel("prices")
plt.title("show Data")
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Housing.csv')
print(df)
```

```python
price = df['price'].tail()
area = df['area'].tail()
```

```python
plt.figure(figsize= (5,5))
plt.plot(price,area)
plt.xlabel("Area of house")
plt.ylabel("price of house")
plt.title("Houceing data")
plt.grid()
plt.show()
```

```python
names  = ['a','s','r','h','y']
marks =[2,3,4,5,6]
```

```python
plt.bar(names,marks)
plt.xlabel("Name of students")
plt.ylabel("marks of students")
plt.title("Data of students")
plt.grid()
plt.legend()
#plt.show()
```

```python
from matplotlib.lines import lineStyles
plt.figure(figsize=(10,8))
plt.scatter(price,area , color ='red', marker = 'o')
plt.xlabel("price of house")
plt.ylabel("area of house")
plt.title("Data of house")
plt.grid()
```

```python
from matplotlib import figure
plt.figure(figsize=(5,5))
plt.hist(df['area'],bins=2,color='red')
plt.xlabel("x axis of area")
plt.ylabel("y axis of area")
plt.title("Data of area")
plt.grid()
plt.legend()
```

```python
df['price'] = df['price'].fillna(df['price'].mean())
```

```python
plt.figure(figsize=(5,5))
plt.boxplot(df['price'],
    patch_artist= True,
    boxprops= dict(facecolor = 'lightblue',color ='blue'),
    whiskerprops=dict(color ="green"),
    capprops=dict(color ="red"),
    medianprops=dict(color ="orange",linewidth = 2),
    flierprops= dict (
        marker = 'o',
        markerfacecolor ="red",
        markersize=10
    )
            )
plt.show()
```

```python
fs = df['furnishingstatus'].value_counts()
plt.pie(fs.values,labels=fs.index,autopct="%1.1f%%")
```

```python
#start seaborn lib.
sns.regplot(x = df['price'],y =df['area'])
```
